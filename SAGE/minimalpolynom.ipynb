```python
def methodone(M):
    
    cp = M.charpoly()
    
    factors = [cp.factor()[i][0] for i in range(len(cp.factor()))]
    multiplicities = [cp.factor()[i][1] for i in range(len(cp.factor()))]
    
    f = reduce(lambda a,b:a*b, factors)
    if f(M) == 0:
        return f
    else:
        olddegrees = [[0 for i in range(len(cp.factor()))]]
        
        for j in range(f.degree(), cp.degree()):
            
            newdegrees = []
            for olddegree in olddegrees:
                
                for k in range(len(olddegree)):
                    
                    if olddegree[k] < multiplicities[k]:
                        newdegree = olddegree.copy()
                        newdegree[k] = newdegree[k] + 1
                        
                        if not (newdegree in newdegrees):
                            g = f
                            
                            for l in range(len(factors)):
                                g = g * factors[l]^newdegree[l]
                                
                            if g(M) == 0:
                                return g
                            else:
                                newdegrees.append(newdegree)
            olddegrees = newdegrees

def methodtwo(M):
    
    F.<x> = PolynomialRing(M.base_ring(),'x')
    
    trackedEntries = [[1,1]]
    
    for j in range(1, M.ncols() + 1):
        W = column_matrix(M.base_ring(), [vector(M.base_ring(),[(M^k)[index[0],index[1]] for index in trackedEntries]) for k in range(j + 1)])
        # sage berechnet linker Kernen...
        Ker = W.transpose().kernel()
        
        if Ker.dimension() == 1:
            vx = matrix(F, j + 1, 1, lambda a, b: x^a)
            v = Ker.gen()
            f = transpose(vx) * v
            evaluation = f[0](M)
            
            if evaluation == 0:
                f = f / f[0][f[0].degree()]
                return f[0]
            else:
                p = 0
                while evaluation[(p - (p % M.ncols())) / M.ncols() , p % M.ncols()] == 0:
                    p = p + 1
                trackedEntries.append([(p - (p % M.ncols())) / M.ncols() , p % M.ncols()])
                            
def methodthree(M):
    
    F.<x> = PolynomialRing(M.base_ring(),'x')
    
    for j in range(1,M.ncols() + 1):
        W = column_matrix(M.base_ring(), [(M^k).list() for k in range(j + 1)])
        Ker = W.transpose().kernel()
        
        if Ker.dimension() == 1:
            vx = matrix(F, j + 1, 1, lambda a, b: x^a)
            v = Ker.gen()
            f = transpose(vx) * v
            
            if f[0](M) == 0:
                f = f / f[0][f[0].degree()]
                return f[0]
```
